```python
#merge
df = pd.merge(left = df_relation, right = bronx_pluto1, left_on = 'Incident Address', right_on = 'Address')
df = (df[df.YearBuilt > 1910])
df.head()
```

```python
data2 = df.dropna(subset=['Incident Zip'],axis=0)
data2.isnull().sum()
```

```python
#As the other NaN are present in columns that will be deleted, there is no need to drop with more NaN numbers

#Define independent variables
xtot=data2.loc[:,data2.columns != "number"]
#Drop unnecesary information for prediction
xtot=xtot.drop(['Unnamed: 0','Unique Key','Created Date','Closed Date','Location Type','Street Name','City',
           'Resolution Description','Address','YCoord','XCoord','Incident Address','Address Type','Status',
            'Borough','BldgClass','Complaint Type'],axis=1)
#print the header to check
xtot.head()
```
